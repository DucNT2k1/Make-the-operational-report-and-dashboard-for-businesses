The values in column 'Trạng thái' are quite similar to column 'Level', they both contain customer status information. So I fill in the null values in this column using the information in column 'Level'.

First, the column 'Level' needs to be processed. It can be seen that at the beginning of each value in the column there are confusing characters such as: L8.1, L8.2,... Therefore, these characters need to be removed.

```python
#Remove characters before '-'
sales_1['Level'] = sales_1['Level'].str.split('-').str[-1]
#Delete the " " character at the beginning
sales_1['Level'] = sales_1['Level'].str.strip()
```

Column 'Level' contain 7 values, including:

*   Khách đã thanh toán toàn bộ (Lead has paid in full)


*   Hủy đơn sau khi chốt (Cancel order)    
*   Không nghe máy/Thuê bao (Not answer the phone/Unable to contact)

*   Băn khoăn về giá (Confused about the price)

*   Trùng    
*   Đã được tư vấn đầy đủ (Fully consulted)

*   Không còn nhu cầu (no longer need to buy products)   

Next, check when column 'Level' contains these values, what values does column 'Trạng thái' contain






         
         
                 
                       
          




```python
#FilterLevel
values_to_keep = ['Khách đã thanh toán toàn bộ','Hủy đơn sau khi chốt','Không nghe máy/Thuê bao','Băn khoăn về giá','Trùng','Đã được tư vấn đầy đủ','Không còn nhu cầu']
#FilterData
Filter_data = sales_1[sales_1['Level'].isin(values_to_keep)]
Filter_data
#GroupBy
groupby = Filter_data.groupby(['Level','Trạng thái'])['Trạng thái'].count()
print(groupby)
```

Fill in the null values in the column 'Status' with the value that appears most often when observing the value pairs of the columns 'Level' and 'Status'.

For example, when the column 'Level' has the value "Hủy đơn dau khi chốt", I will fill in the 'Status' column with the value "Hủy đơn sau khi chốt" because this pair of values appears the most (530 times). Do the same with the remaining rows.


*   Băn khoăn về giá --> Suy nghĩ thêm/Trả lời sau

*   Hủy đơn sau khi chốt --> Hủy đơn sau khi chốt

*   Khách đã thanh toán toàn bộ --> Đặt hàng
*   Không còn nhu cầu --> Từ chối


*   Không nghe máy/Thuê bao --> Không nghe máy


*   Trùng --> Trùng


*   Đã được tư vấn đầy đủ --> Suy nghĩ thêm/Trả lời sau








  

  



  

  


```python
#Fill in null
sales_1['Trạng thái'] = sales_1['Trạng thái'].fillna(sales_1['Trạng thái'].where(~(sales_1['Level'] == 'Băn khoăn về giá'), 'Suy nghĩ thêm/Trả lời sau'))
sales_1['Trạng thái'] = sales_1['Trạng thái'].fillna(sales_1['Trạng thái'].where(~(sales_1['Level'] == 'Hủy đơn sau khi chốt'), 'Hủy đơn sau khi chốt'))
sales_1['Trạng thái'] = sales_1['Trạng thái'].fillna(sales_1['Trạng thái'].where(~(sales_1['Level'] == 'Khách đã thanh toán toàn bộ'), 'Đặt hàng'))
sales_1['Trạng thái'] = sales_1['Trạng thái'].fillna(sales_1['Trạng thái'].where(~(sales_1['Level'] == 'Không còn nhu cầu'), 'Từ chối'))
sales_1['Trạng thái'] = sales_1['Trạng thái'].fillna(sales_1['Trạng thái'].where(~(sales_1['Level'] == 'Không nghe máy/Thuê bao'), 'Không nghe máy'))
sales_1['Trạng thái'] = sales_1['Trạng thái'].fillna(sales_1['Trạng thái'].where(~(sales_1['Level'] == 'Trùng'), 'Trùng'))
sales_1['Trạng thái'] = sales_1['Trạng thái'].fillna(sales_1['Trạng thái'].where(~(sales_1['Level'] == 'Đã được tư vấn đầy đủ'), 'Suy nghĩ thêm/Trả lời sau'))
```
